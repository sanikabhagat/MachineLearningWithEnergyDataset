```python
entities = {
    "energydata" : (df, "id")
}
```

```python
es=ft.EntitySet("entityset", entities)
```

```python
es["energydata"].variables
```

```python
es = es.normalize_entity(base_entity_id="energydata",
                             new_entity_id="lights",
                             index="Appliances",additional_variables=["T2", "T5", "T7"])

```

```python
es["lights"].variables
```

```python
es
```

```python
new_relationship = ft.Relationship(es["lights"]["Appliances"],
                                     es["energydata"]["Appliances"])
```

```python
feature_matrix, feature_defs = ft.dfs(entityset=es,target_entity="energydata", max_depth=3, verbose=1)
```

```python
feature_matrix, feature_defs = ft.dfs(entityset=es,target_entity="lights", max_depth=3, verbose=1)
```
